## 3. ***Data Wrangling***

### Data Wrangling Code

````python
df['Category'] .value_counts()
````

````python
df[df['Category'] == '1.9']
# Correcting a data anomaly at index 10472 using .shift() function
df.loc[10472] = df.loc[10472].shift()
df['App'].loc[10472] = df['Category'].loc[10472]
df['Category'].loc[10472] = 'Lifestyle'
df.loc[10472]
````

````python
df['Rating']=df['Rating'].fillna(0)
df['Size']=df['Size'].fillna('0')
df['Type']=df['Type'].fillna('0')
df['Content Rating']=df['Content Rating'].fillna('0')
df['Current Ver']=df['Current Ver'].fillna('0')
df['Android Ver']=df['Android Ver'].fillna('0')
````

````python
# Write your code to make your dataset analysis ready.
df=df.drop_duplicates(subset='App')
df
````

````python
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
df['Last Updated'][0]
````

````python
df['Last Updated Month']=df['Last Updated'].dt.month
df['Last Updated Year']=df['Last Updated'].dt.year
````

````python
df.drop('Last Updated',axis=1,inplace=True)
````

````python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')
````

````python
df['Rating']=pd.to_numeric(df['Rating'],errors='coerce')
df.loc[df['Rating'] > 5, 'Rating'] = df['Rating'].median()
````

````python
df['Installs'] = df['Installs'].astype(str)
df['Installs'] = df['Installs'].str.replace(',', '')
df['Installs'] = df['Installs'].str.replace('+', '')
df['Installs'] = df['Installs'].astype(int)
````

````python
# Convert all the values in 'size' column to MB
def convert_values(size):
  size = str(size)
  if 'k' in size:
    return float(size.replace('k', ''))
  elif 'M' in size:
    return float(size.replace('M', '')) * 1024
  elif 'Varies with device' in size:
    return np.nan
  return size
````

````python
df['Size']= df['Size'].apply(convert_values)
df['Size']=df['Size'].astype(float)
````

````python
df['Price']=df['Price'].astype(str).str.replace('$','')
df['Price']=pd.to_numeric(df['Price'])
````

##### 2. What is/are the insight(s) found from the chart?

Okay, let's discuss the insights that can be derived from a pair plot:

The insights gained from a pair plot are primarily focused on understanding the relationships between different numerical variables in a dataset. Here are some key insights that can be observed:

1. Correlations: Pair plots can reveal correlations between variables. If two variables show a clear linear pattern in their scatter plot, it suggests a correlation. The strength and direction of the correlation can be inferred from the slope of the line. A positive slope indicates a positive correlation, while a negative slope indicates a negative correlation.

2. Clusters: Pair plots can help identify clusters or groupings of data points. If data points tend to cluster together in specific regions of the scatter plots, it suggests that those data points share similar characteristics. Clusters can be indicative of underlying patterns or relationships within the data.

3. Outliers: Pair plots can help identify outliers, which are data points that are significantly different from other data points in the dataset. Outliers can be spotted as points that are far away from the main cluster of data points in the scatter plots. Identifying outliers is crucial as they can potentially skew the results of statistical analyses.

4. Distribution of Variables: The diagonal plots in a pair plot show the distribution of individual variables. This information is useful for understanding the range, central tendency, and spread of each variable. For example, a histogram with a long tail might suggest that the variable is skewed.

5. Non-linear Relationships: While pair plots are primarily used to identify linear relationships, they can also reveal non-linear relationships between variables. If the scatter plot shows a curved or non-linear pattern, it suggests a more complex relationship between the variables.

By carefully examining the scatter plots, histograms, and density plots within a pair plot, you can gain a deeper understanding of the relationships between variables in your dataset and uncover potential patterns or trends. This information can then be used to guide further analysis or modeling.
